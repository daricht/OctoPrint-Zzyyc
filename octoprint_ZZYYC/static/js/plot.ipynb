```python
import json
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load JSON data
#with open('pointCloud.json', 'r') as f:
with open('pointCloudTidy.json', 'r') as f:
    data = json.load(f)
```

```python

# Extract x, y, and z values
x = [d['x'] for d in data]
y = [d['y'] for d in data]
z = [d['z'] for d in data]

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)

#increase plotsize
fig.set_size_inches(18.5, 10.5)


# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set view to be from above
ax.view_init(elev=90, azim=270)

# Show plot
plt.show()
```

```python
# Create a scatter plot with color-coded z values
plt.scatter(x, y, c=z, cmap='viridis')

# Set the x and y axis limits
plt.xlim([min(x)-1, max(x)+1])
plt.ylim([min(y)-1, max(y)+1])
# plt.xlim([100, 125])
# plt.ylim([70, 90])


#increase plotsize
plt.gcf().set_size_inches(18.5, 10.5)

#set the point size


# Add a color bar to the plot
cbar = plt.colorbar()
cbar.set_label('z')

# Add axis labels and a title to the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('XY plot with z values color-coded')



# Show the plot
plt.show()
```
