```python
# count the number of points in pointcloud

import json
import numpy as np
import matplotlib.pyplot as plt

# load pointcloud
with open('pointCloud.json', 'r') as f:
    data = json.load(f)

#make a list of points with x, y, z coordinates json looks like this: [{"x":0,"y":0,"z":-0.01},{"x":10,"y":0,"z":-0.07},...]
points=[]
for i in range(len(data)):
    points.append([data[i]["x"],data[i]["y"],data[i]["z"]])

# get the number of points
num_points = len(points)

#output the number of points
print("raw data:")
print(num_points)
print(data)
print(points)

#round the x and y coordinates to integers
for i in range(num_points):
    points[i][0] = round(points[i][0])
    points[i][1] = round(points[i][1])

# sort the points: 

# sort by x
points.sort(key=lambda x: x[0])
print("sorted by x:")
print(points)

# sort by y for mathing x
points.sort(key=lambda x: (x[0], x[1]))
print("sorted by y within x:")
print(points)

#fill in missing points
#find the number of points in x
num_points_x = len(set([x[0] for x in points]))
print("number of points in x:")
print(num_points_x)

#find the number of points in y
num_points_y = len(set([x[1] for x in points]))
print("number of points in y:")
print(num_points_y)

#loop through the points and fill in missing points with z=0
for i in range(num_points_x):
    for j in range(num_points_y):
        #print(points[i][0],points[j][1])
        if [i,j] not in [x[0:2] for x in points]:
            points.append([i,j,0])
            #print("added point at: ",i,j)

print("filled in points:")
print(points)


```

```python
# sort by y for mathing x
points.sort(key=lambda x: (x[1], x[0]))
print("sorted by y within x:")
print(points)
num_points = len(points)
print("number of points:")
print(num_points)

lengthx = len(set([x[0] for x in points]))
lengthy = len(set([x[1] for x in points]))
print("length x, y:")
print(lengthx ,lengthy)
maxx = max([x[0] for x in points])
maxy = max([x[1] for x in points])
print("max x, y:")
print(maxx ,maxy)

#make a sorted list of all x coordinates and remove duplicates
x = sorted(list(set([x[0] for x in points])))
print("x:")
print(x)
```

```python

#export the list "points" to a json file
with open('pointCloudTidy.json', 'w') as f:
    json.dump(points, f)
    
```
