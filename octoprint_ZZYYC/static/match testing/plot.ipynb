```python
import json
import matplotlib.pyplot as plt

# Load JSON data
with open('pointCloudTidy.json', 'r') as f:
    data = json.load(f)

# Extract x, y, and z values
x = [d[0] for d in data]
y = [d[1] for d in data]
z = [d[2] for d in data]

# Set threshold for highlighting points
threshold = 21

# Create a scatter plot with color-coded z values
plt.scatter(x, y, c=z, cmap='viridis')

# Highlight points with z values above the threshold in red
plt.scatter([x[i] for i in range(len(z)) if z[i] > threshold],
            [y[i] for i in range(len(z)) if z[i] > threshold],
            c='r', s=50)

# Set the x and y axis limits
plt.xlim([min(x)-1, max(x)+1])
plt.ylim([min(y)-1, max(y)+1])

# Increase plot size
plt.gcf().set_size_inches(18.5, 10.5)

# Add a color bar to the plot
cbar = plt.colorbar()
cbar.set_label('z')

# Add axis labels and a title to the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('XY plot with z values color-coded')

# Show the plot
plt.show()
```

```python
import json
import matplotlib.pyplot as plt

# Load JSON data
with open('pointCloudTidy.json', 'r') as f:
    data = json.load(f)

# Extract x, y, and z values
# x = [d[0] for d in data]
# y = [d[1] for d in data]
# z = [d[2] for d in data]

lower_threshold = -1.2
upper_threshold = 1
x = [d[0] for d in data if lower_threshold <= d[2] <= upper_threshold]
y = [d[1] for d in data if lower_threshold <= d[2] <= upper_threshold]
z = [d[2] for d in data if lower_threshold <= d[2] <= upper_threshold]


# Find the highest x value for each y value, based on z value
highest_x = {}
for i in range(len(x)):
    if y[i] not in highest_x or z[i] > highest_x[y[i]][1]:
        highest_x[y[i]] = (x[i], z[i])

# Create a scatter plot with color-coded z values
plt.scatter(x, y, c=z, cmap='viridis')

# Highlight the highest x value for each y value in red
plt.scatter([highest_x[y[i]][0] for i in range(len(y))],
            y, c='r', s=50)

# Set the x and y axis limits
plt.xlim([min(x)-1, max(x)+1])
plt.ylim([min(y)-1, max(y)+1])

# Increase plot size
plt.gcf().set_size_inches(18.5, 10.5)

# Add a color bar to the plot
cbar = plt.colorbar()
cbar.set_label('z')

# Add axis labels and a title to the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('XY plot with highest z values highlighted')

# Show the plot
plt.show()
```
